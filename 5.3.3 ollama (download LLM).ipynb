# Running a Small Local LLM in Colab (Without Ollama)

This notebook explains why Ollama often fails in Colab and shows a reliable alternative using llama-cpp-python with a GGUF model.

## Why Ollama failed in Colab

Colab environments:
- do not allow persistent background daemons
- reset PATH after installation scripts
- block some system services

Ollama requires a long running service which is not stable in Colab.

## 2) Download a tiny model

```
from huggingface_hub import hf_hub_download

model_path = hf_hub_download(
    repo_id="TheBloke/TinyLlama-1.1B-Chat-v1.0-GGUF",
    filename="tinyllama-1.1b-chat-v1.0.Q4_K_M.gguf"
)

print("Model downloaded to:", model_path)
```

## 3) Load the model

```
from llama_cpp import Llama

llm = Llama(model_path=model_path)
```

## 4) Ask a question

```
def ask(prompt):
    out = llm(prompt, max_tokens=128)
    return out["choices"][0]["text"]

print(ask("Explain RAG in two lines."))
```
